```python
import pandas as pd
import ast

df = pd.read_csv('raw_data/all_video_games(cleaned).csv')
# df[df[['User Score', 'User Ratings Count']].isnull().any(axis=1)]
# Function to safely convert string to Python object
def convert_string_to_dict(platform_info_str):
    try:
        return ast.literal_eval(platform_info_str)
    except (ValueError, SyntaxError):
        return None  # Return None or a custom value if parsing fails

# Apply the function to the entire 'Platforms Info' column
df['Platforms Info'] = df['Platforms Info'].apply(convert_string_to_dict)
```
