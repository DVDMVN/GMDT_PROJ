```python
# Loading Datasets from kaggle
import kaggle
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from utils import horizontal

kaggle.api.authenticate()  # To use the api, a kaggle.json file is necessary for authentication information. For more info: https://www.kaggle.com/docs/api
if not os.path.exists('data'):
    kaggle.api.dataset_download_files(  # Downloads a dataset given the slug for owner/dataset_name, unzip=True unzips files on download, path is specified to data
        "alicemtopcu/top-1500-games-on-steam-by-revenue-09-09-2024", unzip=True, path="data"
    )
    kaggle.api.dataset_download_files(
        "tristan581/all-55000-games-on-steam-november-2022", unzip=True, path="data"
    )
    kaggle.api.dataset_download_files(
        "beridzeg45/video-games", unzip=True, path="data"
    )

top_1500_steam = pd.read_csv("data/Steam_2024_bestRevenue_1500.csv")
all_55000_steam = pd.read_csv("data/steam_games.csv", sep=';', low_memory=False)
metacritic_reviews = pd.read_csv("data/all_video_games(cleaned).csv")
```

```python
# Displaying data basics
def display_basics(df: pd.DataFrame):
    display(df.head())
    object_cols = df.select_dtypes(include=['object']).columns
    horizontal([df.describe(), df[object_cols].describe()])

display_basics(top_1500_steam)
display_basics(all_55000_steam)
display_basics(metacritic_reviews)
```
