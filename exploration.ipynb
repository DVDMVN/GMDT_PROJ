```python
# Downloading the data

import kaggle

kaggle.api.authenticate()  # To use the api, a kaggle.json file is necessary for authentication information. For more info: https://www.kaggle.com/docs/api

kaggle.api.dataset_download_files(  # Downloads a dataset given the slug for owner/dataset_name, unzip=True unzips files on download, path is specified to ./mushroom_data
    "alicemtopcu/top-1500-games-on-steam-by-revenue-09-09-2024", unzip=True, path="top_steam_data"
)
```

```python
# Reading into DataFrame
import pandas as pd
from IPython.display import HTML

top_steam_games = pd.read_csv('top_steam_data/Steam_2024_bestRevenue_1500.csv')

# Borrowing tidier from previous assignment
def horizontal(dfs):
    html = '<div style="display:flex">'
    for df in dfs:
        html += '<div style="margin-right: 32px">'
        html += df.to_html()
        html += "</div>"
    html += "</div>"
    display(HTML(html))

display(top_steam_games.head())
object_cols = top_steam_games.select_dtypes(include=['object']).columns
horizontal([top_steam_games.describe(), top_steam_games[object_cols].describe()])
```
